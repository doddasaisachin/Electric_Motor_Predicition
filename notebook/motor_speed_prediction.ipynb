```python
def correlated(dataframe):
    relations=[]
    corr_df=dataframe.corr()
    for i in dataframe.columns:
        for j in dataframe.columns:
            if i!=j:
                if (corr_df[i][j]>=0.90) or (corr_df[i][j]<=-0.90):
                    if set((i,j)) not in relations:
                        relations.append(set((i,j)))
    return relations
```
